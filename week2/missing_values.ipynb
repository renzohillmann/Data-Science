# Method paramete()

bfill and ffill are the two methods that are used to fill the missing values in the dataframe. Both of these methods are used to fill the missing values in the dataframe. The bfill() function is used to backward fill the missing values in the dataframe. The ffill() function is used to forward fill the missing values in the dataframe. However, we need to sort the data before using these functions.

```python
df = df.set_index('time')
df = df.sort_index()

df.head(20)
```

```python
df.index
```

```python
# We can see that the time isn't really unique

df = df.reset_index() # reset the index because we previously promoted the time to an index.
df = df.set_index(["time", "user"])

df.head(20)

```

```python
df = df.fillna(method = 'ffill')
df.head(20)
```

# We can create custom ffills/bfills by using the .replace() method

```python
# let's create a new DataFrame

df = pd.DataFrame({'A': [1, 1, 2, 3, 100], 'B': [3, 4, 5, 6, 8], 'C': [10, 20, 30, 40, 50]})

df

```

```python
df.replace(100, 4)
```

```python
# changing two values at the same time. THE ORIGINAL DATAFRAME IS CHANGED

df.replace([1, 100], [50, 1])
```

# The replace method also supports regex

```python
df = pd.read_csv("log.csv")
df.head(20)
```

# THE FOLLOWING SOLUTION DOESN'T WORK IN MY LOCAL MACHINE

NO IT WORKS, BECAUSE I WAS GETTING A NEW DATAFRAME INSTEAD OF MODIFYING THE ORIGINAL DATAFRAME

```python
# I need to use df = df.replace because the replace method returns a new DataFrame

df = df.replace(to_replace=".*.html$", value="webpage", regex=True)

df.head(20)
```

```python
# I can also specify the column to perform the replacement

df["video"] = df["video"].replace(to_replace=".*.html$", value="webpage", regex=True)
df.head
```

# The following code WORKS!

```python
for col in df.columns:
    df[col] = df[col].replace(to_replace=".*.html$", value="webpage", regex=True)

df.head(20)
```
