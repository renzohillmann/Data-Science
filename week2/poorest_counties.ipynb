# Comparing the Poorest and Richest Counties in the United States

The two poorest counties in the United States are Crowley County, Colorado and Wheeler County, Georgia

The two richest counties in the United States are Loudoun County, Virginia and Fairfax County, Virginia

We want to know their populations.

```python
import pandas as pd

df = pd.read_csv("census.csv")

df
```

```python
# We just want to query the populations of the counties.
df = df[df['SUMLEV'] == 50]
```

```python
columns_to_keep = ['STNAME', 'CTYNAME', 'CENSUS2010POP']

df = df[columns_to_keep]
```

```python
#SETTING THE DOUBLE INDEX

df = df.set_index(['STNAME', 'CTYNAME'])
df.head()
```

```python
# querying all the counties in Colorado and saving it into a txt file

df.loc['Colorado'].to_csv('colorado_counties.txt', sep='\t')
print(df.loc['Colorado'])
```

```python
# finding the populations of poorest counties:

print(df.loc[('Colorado', 'Crowley County')])
print(df.loc[('Georgia', 'Wheeler County')])

```

```python
# printing Virginia counties to a txt file

df.loc['Virginia'].to_csv('virginia_counties.txt', sep='\t')
```

```python
# finding the populations of the richest counties:

print(df.loc[('Virginia', 'Loudoun County')])
print(df.loc[('Virginia', 'Fairfax County')])
```
