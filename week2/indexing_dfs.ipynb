# Multiple indexing

```python
df = pd.read_csv("census.csv")
df.head()
```

```python
#Similar to the SQL DISTINCT operator.

df["SUMLEV"].unique()
```

Let's exclude all the rows thata are summaries at the stated level and just keep the ones at the county data.

```python
df = df[df["SUMLEV"] == 50]
df.head()
```

Let's reduce the data we will be looking at to just the total population estimates and the total number of births.

```python
columns_to_keep = ["STNAME", "CTYNAME", "BIRTHS2010", "BIRTHS2011", "BIRTHS2012", "BIRTHS2013", "BIRTHS2014", "BIRTHS2015", "POPESTIMATE2010", "POPESTIMATE2011", "POPESTIMATE2012", "POPESTIMATE2013", "POPESTIMATE2014", "POPESTIMATE2015"]

df = df[columns_to_keep]    
df.head()
```

```python
# Which states we have in our data set

df["STNAME"].unique()
```

```python
df[df["STNAME"] == "Virginia"]
```

The US Census data breaks down population estimates by state and county. We can set the index as a combination of state and county to get a unique index for each row.

```python
df = df.set_index(["STNAME", "CTYNAME"])
df
```

# How to query dataframes with multi-indexes

You must provide the arguments in order of the index levels.

```python
# Population of the Washteaw County in Michigan

df.loc[("Michigan", "Washtenaw County")]
```

```python
# Comparing two counties. We must provide a list of tuples

df.loc[ [("Michigan", "Washtenaw County"), ("Michigan", "Wayne County")] ]
```

```python
df.loc[ [("Virginia", "Loudon County"), ("Colorado", "Crowley County")]]
```
