# Boolean masking

This is the key concept here. We can imagine a Boolean masking as an array in top of our dataframe or series full of Trues of Falses, depending on the condition we want to include in our final result.

```python
import pandas as pd

df = pd.read_csv("Admission_Predict.csv", index_col= 0)

df.columns = [x.strip().lower() for x in df.columns]

df.head()
```

```python
admit_mask = df["chance of admit"] > 0.7
admit_mask # The result is a series
```

```python
df.where(admit_mask).head() #The original dataframe is not modified
```

```python
#we can drop the NaNs

df.where(admit_mask).dropna().head()
```

Ther is a shorthand that includes both where and dropna

```python
df[df["chance of admit" ] > 0.7].head()
```

## The indexing operator does now three things:

Each of those are mimicking functionalities of .iloc and .loc

```python
df["gre score"].head() #returns a series
```

```python
# we can pass a list of columns

df[["gre score", "toefl score"]].head()
```

```python
# we can pass a masking

df[df["gre score"] > 330].head()
```

# Combining multiple Boolean maskings

```python
(df["chance of admit"] > 0.7) & (df["chance of admit"] < 0.9) #we need to add the parenthesis () & () to not get an error
```

# We can use built in functions

```python
df["chance of admit"].gt(0.7) & df["chance of admit"].lt(0.9)
```

```python
# We can chain these operations

df["chance of admit"].lt(0.9).gt(0.7)

#Why this give a different answer.

# This is incorrect!!! WE CANNOT CHAIN THEM!!! Python interprets True and False as 1 and 0 as int datatype and compare them
```
